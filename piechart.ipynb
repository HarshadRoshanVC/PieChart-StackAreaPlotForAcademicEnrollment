```python
#importing necessary packages
import pandas as pd
import matplotlib.pyplot as plt

#reading enrollment.csv file
enrollment = pd.read_csv('enrollment.csv',index_col='Academic Unit')
#plotting first pie chart
plot = enrollment.plot.pie(y='Enrollment')
```

```python
#sorting and taking first 5 academic unit
enrollment = enrollment.sort_values('Enrollment',ascending=False).head()
#printng data
print(enrollment)
```

```python
#pie chart without legend and axis label
plot = enrollment.plot.pie(y='Enrollment',legend=False)
ax = plot
ax.set_ylabel('')
```

```python
#pie chart with percentage value
plot = enrollment.plot.pie(y='Enrollment',legend=False,autopct='%1.2f%%',title='SBU Academic Unit Enrollment')
ax = plot
ax.set_ylabel('')
```

```python
#reading enrollment dataset of 5 years
enrollment5year = pd.read_csv('enrollment_5years.csv',index_col='Academic Unit')

#printing details of single row
print(enrollment5year.loc['College of Engineering & App Sci'])
```

```python
#plotting final stack area plot
#Gathering details of College of Arts & science,College of Engineering & App Sci,College of Business
y1=enrollment5year.loc['College of Arts & science']
y2=enrollment5year.loc['College of Engineering & App Sci']
y3=enrollment5year.loc['College of Business']
#creating list for x labels(xticks) for stack area plot
x=['Fall 2015','Fall 2016','Fall 2017','Fall 2018','Fall 2019']
plt.stackplot(x,y1,y2,y3,labels=('College of Arts & science','College of Engineering & App Sci','College of Business'))
#setting y axis label
plt.ylabel('Enrollment')
#putting legend outside plot
plt.legend(loc='upper left',bbox_to_anchor=(1.0,1.0))
#giving title
plt.title('SBU Academic Unit Enrollment')
plt.show()
```
